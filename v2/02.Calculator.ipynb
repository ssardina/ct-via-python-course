<font size="6" color="#b21200"  face="verdana"> <B>Basic Math Operations</B></font>

As you can imagine, Python will b able to do the typical operations of addition (`+`), subtraction (`-`), multiplication (`*`) and division (`/`). 

Check some examples and pay attention to the comments on each line:

````python
print(23) # justs print the integer number 23
print(23 + 5) # addition
print(23 - 5) # substraction
print(23 * 5) # multiplication
print(23 / 2) # division
````

Notice that whenever Python finds an expression, like "`23 + 5`", it will print the _result_ of the expression (afer calculating it of course).

Note also that all numbers above, like `23` and `5`, are number without decimal points; they are called **integers** (that is, a whole number). 

You can also use numbers with decimal points, known as **floats**. The above division `23 / 2` result in a float (`11.5`), not a integer.

Check a similar sequence of five calculations using floats:

````python
print(23.50) # justs print the float number 23.50
print(23.50 + 5.3) # addition
print(23.50 - 5.3) # substraction
print(23.50 * 5.2) # multiplication
print(23.50 / 4.1) # division
````

Of course, Python can do much more than the classical four math operations. Consider the following ones:

````python
print(22 // 5) # floor division, discards the decimal part
print(22 % 5) # reminder of the division 22 / 5
print(3 ** 4) # exponentiation, 3 rised to the power of 4 (i.e., 3 mutiplied 4 times: 3*3*3*3)
````

## Complex Expressions

All the above calculations have just one operator each, as was the case in old simple calculators. But Python is much more powerful and versatile, so it can handle huge expressions using many operators! For example, try running this code:

````python
print((5 + 2) * 24) # How many hours there is in a week of 5 working days and 2 weekend days
````

But, why did we use so many parenthesis? Well, the outer ones correspond to the `print` statement/function, so they are not part of the expression. What about the ones around `5+2`? Let's remove it and see what is happening:

````python
print(5 + 2 * 24) 
````

Ouch! So, seems that `(5 + 2) * 24` and `5 + 2 * 24` are not the same for Python.

The reasons is that, in the abscence of explicit `(...)` operators have different precendece_ levels. For example, in math multiplication (`*`) has higher precendence than addition (`+`), so in the latest calculation above, `2 * 24` are done first, and then `5` is added. So, without parenthesis, the expresion is the same as writting:

````python
print(5 + (2 * 24))
````

The common **precedence level** is: exponentiation, multiplication, reminder, division or floor division, addition, substraction. 

The advice is to always use parenthesis when in doubt, so as to make sure you the computer is doing the calculation you really intended. Once you start getting familiar with the precedence of math operators you will learn how to use them only when really necessary (as in the above example!)

<font size="5" color="#b21200"  face="verdana"> <B>Numbers vs Strings</B></font>


In the previous notebook we have seen strings: lists of characters to represent any text.

You may be wondering why in this notebook we have not used the usual quotes (`"` or `'`) when printing. Here is why:

````python
print("22 + 8")
print('22 + 8')
````

Python didn't print `30`! _Can you guess what is happening?_ 

The fact is that `"22 +8"` or `'22+8'` are _not_ mathematical expressions, they are just strings! So, Python prints them _as they are_ without evaluating anything at all. Python doesn't even know (or care) what is within the quotes. One can even write wrong expressions, and Python will just print they verbatim:

````python
print("22 +- 8**hello world")
````

See? If a number is inside the quotes, it is not treated as a number (integer or float), it is treated as just text.

## Incrementing and Decrementing Variables

Consider a video game sofware. We will often use a variable, say `player_score`, to store the current score of the player.  Initially, when th game starts, the score is zero:

````python
player_score = 0 # the player_score variable is set to 0 at start of game

print("Game started. Player score is:", player_score)
````

As the game progresses, the score of the player keeps changing, up and down, depending how good/bad the player performs. So, for example, at some point the player will earn one more point, so the video game has to **update the `player_score` variable**. It does so, as follows: 

````python
player_score = player_score + 1   # add one to player_score and store back new result in player_score itself!

print("Congrats, you earn +1 extra points! New score is:", player_score)
````

### Programming variables vs Math variables

While programing variables have a lot in common with the usual math variables that you see in algebra (like `x + 2 = y`), they are not exactly the same:

1. Variables in code can change their values, as we have seen above with `player_score` as they just mean a location in memory where some value can be stored. Variable in math are not locations in memory and cannot "change" their value: variable `x` can only be one number.

2. The `=` symbol above is not the equal sign as in math. So, the first line above **does not** mean that `player_score` is equal to itself plus one (impossible! two cannot be equal to three, three to four, etc.). Instead the `=` symbol is the **assignment** operation: it assigns/set the variable on the left hand side (`player_score` in this case) to the result of the right hand side (`player_score + 1`).

<font size="6" color="#b21200"  face="verdana"> <B>Asking numbers to the user</B></font>

So far, we have known what the numbers were for our problems. But our programs should work for any numbers. Things get much more interesting when we ask the user for some numbers, and we then perform some cool calculations.


![ask](../imgs/ask-2341784_640.resized.jpg)

````python
n = input('Pick a number you like: ')     # ask the user for a number
print("Your number is", n)         # now print it!
````

Let's now print the double of the number entered:


````python
n = input('Pick a number you like: ')     # ask the user for a number
print("The double of your number is", n + n)         
````

Ops! What has happned here? Well, the function `input(.)` ask the user for some **input as text**, which means that variable `n` is not a number, but a string! So, if the user enters `45`, then variable `n` is actually the string `"45"`. This means that, as we saw above,  the expression `n + n` says to concatenate string `n` with itself, thus producing the intended result.

Fortunately, Python provides a function `int(.)` to **convert** a string to its number representation. So, we first read the user input and then convert that string to its corresponding number:

````python
n_string = input('Pick a number you like: ')     # ask the user for a number
n = int(n_string)
print("The double of your number is", n + n)         
````

Here, the second line says: "convert the string `n_string` to a number (integer or float, as needed) and assign it to variable `n`". We can avoid intermediate steps and just apply the conversion just to the result of the input:


````python
n = int(input('Pick a number you like: '))     # ask the user for a number
print("The double of your number is", n + n)         
````
