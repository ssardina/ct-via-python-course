<font size="5" color="#b21200"  face="verdana"> <B>Sorting Numbers using Bubble Sort</B></font>

Keeping a list _sorted_ is very covenient. Imagine a list of numbers, people in your moble contact list, electoral roll, list of country names, etc. the Can you think why it is useful to have the list "sorted"? 

Many times we have a list that is _unsorted_, and our goal is to sort it. When the list has 5 or 10 elements, it is easy. But what if the list has 100, 1000 or even 1 million numbers or names to be sorted?

Many clever algorithms have been devised for putting values into order efficiently. This activity will explore the [Bubble Sort](https://en.wikipedia.org/wiki/Bubble_sort) algorithm. It can be used to show almost every ingredient of modern programming:

1. An **array** or list of numbers as a data structure.
2. The use of a **variable** to point to a "current" pair of consecutive numbers in the array.
3. The use of the pointer variable as a "**counter**", and assignments `n = n + 1`.
3. **Conditional** `if-then-else` statements when comparing the current pair.
4. **Loop** `while-do` statements to go over the array many times until fully sorted.

You can check the [unplugged bubble sort](https://minecraft.makecode.com/courses/csintro/arrays/unplugged-2) activity from Minecraft, but the idea is simple.

[![Bubble Sort Video Explanation](https://img.youtube.com/vi/SA8ASQ_5H6w/0.jpg)](https://www.youtube.com/watch?time_continue=4&v=SA8ASQ_5H6w "Bubble Sort")

All that is needed is a set of cards with numbers, say from 1 to 100. You can get & print different types and sizes of such cards [here](http://www.sparklebox.co.uk/maths/numbers/flash-cards.html)


1. A number of volunteers, each with a random number, are set up in a line, from left to right, say. They form a **list** or **array** of numbers.
2. Each person has its **assigned** number card hidden.
3. Each person has a "position" in the list: the first position is position 0, the second is position 1, and so on. Arrays in a computer start at zero!
3. Another volunteer acts as the "**pointer**". He/she starts from position 0, that is, from the first person in line.
4. At any point, the pointer volunteer is "pointing" to one person, that is, to one position in the array. Say it is pointing to position `n`.
4. At this point, the people in positions `n` and `n+1` show and compare their numbers (all other cards remain still and hidden):
5. **If the card at position `n` is greater than the one in position `n+1`**, then the two people swap positions. Otherwise, they say "hi" and remain where they are each.
5. The two people who just compared, hide their cards again.
6. After the **comparison**, the pointer volunteer takes a step forward and advances to position `n+1`.
7. If the pointer voluteer reaches the last person, then it walks back to the person at position 0 and starts the process again.
8. The whole process is **repeated** as many times as people in the list minus one. So, if there are 9 people in the list, the process has to be repeated 9 times.
9. At the end, the list is guaranteeed to be sorted. **WHY?**

## Ways to run the activity

There are at least **two ways** one can run this unplugged activity:

1. The whole line up is set-up and the pointer volunteer tries to come up with a way of sorting the list. The rules are simple: she/can can be at one position, see only sees the pair in that position, and can only swap those two people (who are adjacent to each other). If there are too many numbers/volunteers it will be tricky without coming with a strategy. The volunteer may get help from the audience.
2. The algorithm is explained beforehand to everyone, and the pointer volunteer just executes it.
